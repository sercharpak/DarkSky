Generates mock data to test the FOF

```python
def sphere(n_points, x0=0, y0=0, z0=0):
    phi = 2.0 * pi * random.random(n_points)
    theta = arccos((random.random(n_points) - 0.5)*2.0)
    r = random.random(n_points)
    x = r * cos(phi) * sin(theta) + x0
    y = r * sin(phi) * sin(theta) + y0
    z = r * cos(theta) + z0
    return x, y, z 
```

```python
x,y,z=sphere(1000)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, -z, zdir='z', c= 'red')
```

```python
#generates data for 1000 spheres
fileout = open('../src/test_input.dat', 'w')

n_points_total = 10 * 10 * 10 * 100
fileout.write("%d\n"%n_points_total) #points in total 
fileout.write("%d\n"%n_points_total) #points in 'DM'
fileout.write("0\n") #gas
fileout.write("0\n") #stars
fileout.write("0.01\n") # time
fileout.write("0\n") # nactive

center_x = linspace(1.0,20.0,10)
center_y = linspace(1.0,20.0,10)
center_z = linspace(1.0,20.0,10)

for x0 in center_x:
    for y0 in center_y:
        for z0 in center_z:
            x,y,z=sphere(100, x0=x0, y0=y0, z0=z0)
            for i in range(size(x)):
                fileout.write("%f %f %f\n"%(x[i],y[i],z[i]))
fileout.close()
            
```
